# Исследование закономерностей определяющих успех игр

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Описание данных:
- `Name` — название игры;
- `Platform` — платформа;
- `Year_of_Release` — год выпуска;
- `Genre` — жанр игры;
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий);
- `EU_sales` — продажи в Европе (миллионы проданных копий);
- `JP_sales` — продажи в Японии (миллионы проданных копий);
- `Other_sales` — продажи в других странах (миллионы проданных копий);
- `Critic_Score` — оценка критиков (максимум 100);
- `User_Score` — оценка пользователей (максимум 10);
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Предобработка данных

Заменим названия столбцов (приведём их к нижнему регистру)

```python
data.columns = data.columns.str.lower()
```

```python
data.columns
```

```python
# выводим общую информацию по данным
data.info()
```

Преобразуем типы данных в столбцах *year_of_release*, *critic_score*, *user_score*, также заменим пропуски на условный маркер **-1**. В данном случае будет не разумно заменять их на медианные (либо средние) значения, поскольку это испортит дальнейший анализ.

```python
# int т.к. целочисленные значение
data['year_of_release'] = data['year_of_release'].fillna(-1).astype('int')
data['critic_score'] = data['critic_score'].fillna(-1).astype('int')
```

Отдельно разберем столбец *user_score* и значение 'tbd'. tbd расшифровывается как 'to be determined' и означает, что оценки на данный момент нет. Таким образом можно также заменить эти значения на **-1** и изменить тип данных на вещественный *float*.

```python
data[(data['user_score']=='tbd')]
```

```python
# float 64, т.к. содержатся вещественные числа
data.loc[data['user_score']=='tbd', 'user_score'] = -1
data['user_score'] = data['user_score'].fillna(-1).astype('float64')
```

```python
data.head(20)
```

```python
# проверим оставшиеся пропуски
data.isna().sum()
```

Т.к. в столбцах *name* и *genre* всего по два пропуска, то такие строки можно убрать и это не повлияет на результат анализа.

```python
data = data.dropna(subset=['name', 'genre'])
```

Заменим пропуски в столбце *rating* на условный маркер **-1**.

```python
data.loc[data['rating'].isna(), 'rating'] = -1 
```

### Суммарные продажи по регионам

Посчитаем суммарные продажи каждой игры по всем регионам и запишем эти данные в столбец *total_sales*.

```python
# складываем продажи по каждому региону
data['total_sales'] = \
data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
data.head(20)
```

## Анализ данных

### Сколько игр выпускалось в разные годы

Посмотрим сколько игр выпускалось в разные годы. Для начала выведем уникальные года из столбца *year_of_release*

```python
# выводим все года по порядку
data['year_of_release'].sort_values().unique()
```

Построим линейный график по которому можно проследить количество выпущенных игр с 1980 по 2016.

```python
# группируем данные по годам, считаем количество игр, строим график
(
    data
    .groupby('year_of_release')['year_of_release'].count()
    .plot(xlim=(1980,2016), xlabel='год выпуска', ylabel='количество игр', figsize=(12,8))
)
plt.title('Количество игр в год')
plt.show()
```

**Вывод:** игр стали больше делать начиная с 1999 года и пик наблюдается с 2007 по 2009 включительно. При этом в период с 1980 по 1990 игр выходило минимальное количество.

### Продажи игр по платформам

Рассмотрим платформы на которых было продано больше всего игр и сделаем распределение с наибольшими продажами по этим платформам.

```python
# посчитаем количество проданных игр по каждой платформе и выведем график
platforms_sales = data.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
```

```python
platforms_sales.plot(kind='bar', figsize=(13,6), ylabel='продажи, млн', xlabel='платформа')
plt.xticks(rotation=45)
plt.title('Продажи по каждой платформе')
plt.show()
```

Построим динамику по годам для 10 наиболее популярных платформ.

```python
# первый вариант где выводится каждый график отдельно

for platform in platforms_sales.index[:10]:
    data[
        (data["platform"] == platform) & (data["year_of_release"] != -1)
    ]\
        .groupby("year_of_release")["total_sales"].sum()\
        .plot(title=platform, figsize=(8, 4), ylabel="продажи, млн", xlabel="год")
    plt.show()
```

На отдельных графиках наглядно можно проследить срок жизни и пик продаж платформ.

**Вывод:** в среднем платформы достигают своих пиковых продаж в течение 4-5 лет, а дальше идёт спад и они остаются актуальными примерно 3 года. Итого срок жизни платформы 7-8 лет.

### Период актуальности платформы

Учитывая выводы из предыдущих задач, можно сказать что актуальными можно считать платформы, вышедшие за последние 5 лет.

```python
relevant_data = data.query('year_of_release >= 2013')
```

Посчитаем статистику продаж по платформам за актуальный период

```python
relevant_platforms_sales = relevant_data.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
relevant_platforms_sales.head(20)
```

Выводим графики продаж по 10 самым популярным актуальным платформам.

```python
# второй вариант с использованием мульти-индексов и общим графиком
# для большей наглядности

# fig, ax = plt.subplots()

#создаем мульти-индекс (платформа-года)
index_array = []
for platform in relevant_platforms_sales.index[:10]:
    index_array += [[platform, year] for year in range(2013, 2017)]

mul_index = pd.MultiIndex.from_frame(pd.DataFrame(index_array))

#группируем данные и для каждой платформы строим отдельную линию продаж
group_data = relevant_data.groupby(['platform', 'year_of_release'])['total_sales']\
    .sum().sort_values(ascending=False).reindex(mul_index).fillna(-1)

for platform in relevant_platforms_sales.index[:10]:
    group_data[platform].plot(title=platform,
                    xlabel="год", ylabel="продажи", figsize=(8, 6), ylim=[0, 120])
    plt.show()
```

**Вывод:** исходя из полученного графика потенциально прибыльными актуальными платформами можно считать PS4, Xbox One, 3DS, WiiU и PC. PS3 и Xbox 360 несмотря на огромные продажи с 2011 года сильно идут на спад к 2016 году поэтому нельзя сказать что они окажутся потенциально прибыльными, учитывая успех консолей нового поколения (PS4 и Xbox One).

**Дополнение к выводу:** помимо плафторм указанных выше также стоит добавить PSV и тут стоит сделать акцент на том, что это актуальная портативная консоль на равне с 3DS, но продаж по сравнению с ней гораздо меньше.

### Влияние на продажи отзывов пользователей и критиков

Чтобы определить влияние отзывов пользователей и критиков на продажи внутри одной популярной платформы построим диаграмму рассеяния и найдем корреляцию. Для примера возьмем платформу PS4, т.к. у неё самые хорошие показатели по продажам.

```python
# (
#     relevant_data
#     .loc[(relevant_data['platform']=='PS4') & (relevant_data['user_score'] != -1)]
#     .plot(kind='scatter', x='total_sales' , y='user_score', alpha=0.5)
# )
# plt.title('Scatter PS4')
# plt.show()
```

```python
fig, ax = plt.subplots(1, 2)

PS4_sample = \
relevant_data.loc[(relevant_data['platform']=='PS4') & 
                  (relevant_data['user_score'] != -1) &
                  (relevant_data['critic_score'] != -1)]

ax[0].scatter(x=PS4_sample['total_sales'], 
              y=PS4_sample['user_score'], 
              alpha=0.5)
ax[0].set(xlabel='общее число продаж, млн', ylabel='оценка пользователей')

ax[1].scatter(x=PS4_sample['total_sales'], 
              y=PS4_sample['critic_score'], 
              alpha=0.5)
ax[1].set(xlabel='общее число продаж, млн', ylabel='оценка критиков')

fig.suptitle('Диаграмма рассеивания PS4')
fig.set_figheight(5)
fig.set_figwidth(9)
```

**Вывод:** из графиков видно, что точек становится гораздо больше, когда пользователи ставят оценку от 5.5 до 9, либо когда критики ставят от 60 до 90. Из этого можно сделать вывод что хорошие оценки влияют на большие продажи. Выведем также корреляцию общих продаж отдельно по оценкам пользователей и критиков.

```python
print('Корреляция для всех продаж и оценкой пользователя составляет:', 
      PS4_sample['total_sales'].corr(PS4_sample['user_score']))
print('Корреляция для всех продаж и оценкой критика составляет:', 
      PS4_sample['total_sales'].corr(PS4_sample['critic_score']))
```

**Вывод:** при этом корреляция между продажами и оценкой пользователя показывает, что связи практически нет (значение очень близко к нулю -0.03). А оценка критика действительно имеет своё влияние на хорошие продажи, т.к. корреляция принимает значение 0.4 .

Посмотрим общую статистику по всем платформам. Построим также диаграмму рассеивания и посчитаем корреляцию.

```python
fig, ax = plt.subplots(1, 2)

all_platforms = \
relevant_data.loc[(relevant_data['user_score'] != -1) &
                  (relevant_data['critic_score'] != -1)]

ax[0].scatter(x=all_platforms['total_sales'], 
              y=all_platforms['user_score'], 
              alpha=0.2)
ax[0].set(xlabel='общее число продаж, млн', ylabel='оценка пользователей')

ax[1].scatter(x=all_platforms['total_sales'], 
              y=all_platforms['critic_score'], 
              alpha=0.2)
ax[1].set(xlabel='общее число продаж, млн', ylabel='оценка критиков')

fig.suptitle('Диаграмма размаха для всех платформ')
fig.set_figheight(5)
fig.set_figwidth(9)

print('Корреляция для всех продаж и оценкой пользователя составляет:', 
      all_platforms['total_sales'].corr(all_platforms['user_score']))
print('Корреляция для всех продаж и оценкой критика составляет:', 
      all_platforms['total_sales'].corr(all_platforms['critic_score']))
```

**Вывод по всем платформам:** тенденция сохраняется примерно на том же уровне (как для PS4) и для всех платформ, а именно:
- корреляция между продажами и оценкой пользователя показывает, что связи практически нет (близка к нулю);
- оценка критика может влиять на хорошие продажи, т.к. корреляция принимает значение 0.32;
- диаграмма рассеивания для продаж и оценки пользователя отображает большое скопление точек от 5.5 до 9;
- диаграмма рассеивания для продаж и оценки критика отображает большое скопление точек от 60 до 90.

## Портрет пользователя каждого региона

Определим интересующие параметры для пользователя каждого региона (NA, EU, JP). Начнем с топ-5 самых популярных платформ.

```python
for region in ['na_sales', 'eu_sales', 'jp_sales']:
    
    top5_platform =\
    relevant_data.groupby('platform')[region].sum().sort_values(ascending=False)[:5]
    print(f'Топ-5 платформ по региону {region}:\n {top5_platform}\n')
    top5_platform.plot(kind='pie')
    plt.show()
```

## Проверка гипотез

Формулируем нулевую и альтернативную гипотезу, чтобы знак равенства присутствовал в нулевой гипотезе. Обе гипотезы будут двусторонними.

### Пользовательские рейтинги Xbox One и PC

Проверим гипотезу "средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

- H0: средний пользовательский рейтинг Xbox One равен среднему пользовательскому рейтингу PC;
- H1: средний пользовательский рейтинг Xbox One не равен среднему пользовательскому рейтингу PC.

<div class="alert alert-success">

**Артём:** 👍 Гипотезы сформулированы верно.

</div>

Возьмем alpha 0.01, т.к. нам не нужно строгое соответствие чтобы утверждать что средние пользовательские рейтинги равны.

Мы имеем две выборки, поэтому используем метод ttest_ind из библиотеки stats. Параметр equal_var оставляем по умолчанию True, т.к. у нас нет оснований утверждать, что дисперсии разные.

```python
xone_user_score = relevant_data[(relevant_data['platform']=='XOne') & 
                                (relevant_data['user_score']!= -1)]['user_score']
pc_user_score = relevant_data[(relevant_data['platform']=='PC') & 
                                (relevant_data['user_score']!= -1)]['user_score']

alpha = .01
results = st.ttest_ind(xone_user_score, pc_user_score)

print('pvalue:',results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получится отвергнуть нулевую гипотезу')
```

**Вывод:** не получилось отвергнуть нулевую гипотезу, исходя из полученных результатов. Мы не можем утверждать, что средние пользовательские рейтинги Xbox One и PC не равны.

### Рейтинги жанров Action и Sports

Проверим гипотезу "cредние пользовательские рейтинги жанров Action и Sports разные".

- H0: средний пользовательский рейтинг Action равен среднему пользовательскому рейтингу Sports;
- H1: средний пользовательский рейтинг Action не равен среднему пользовательскому рейтингу Sports. 

Возьмем alpha 0.01, т.к. нам не нужно строгое соответствие чтобы утверждать что средние пользовательские рейтинги равны.

Мы имеем две выборки, поэтому используем метод ttest_ind из библиотеки stats. Параметр equal_var оставляем по умолчанию True, т.к. у нас нет оснований утверждать, что дисперсии разные.

```python
action_user_score = relevant_data[(relevant_data['genre']=='Action') & 
                                (relevant_data['user_score']!= -1)]['user_score']
sports_user_score = relevant_data[(relevant_data['genre']=='Sports') & 
                                (relevant_data['user_score']!= -1)]['user_score']

alpha = .01
results = st.ttest_ind(action_user_score, sports_user_score)

print('pvalue:',results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получится отвергнуть нулевую гипотезу')
```

**Вывод:** отвергаем нулевую гипотезу, следовательно можно утверждать, что средние рейтинги жанров Action и Sports разные.
