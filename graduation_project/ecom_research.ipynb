## Подготовка данных

Приведем столбец *date* к типу *datetime*. Также можно изменить тип данных столбца *price* на *int64*.

````python
transactions['date'] = pd.to_datetime(transactions['date'],
               format='%Y%m%d%H')

transactions['price'] = transactions['price'].astype('int')
````

Приведем названия товаров к общему виду, заменим буквы на строчные, а также букву *"ё"* на *"е"*.

````python
transactions['product'] = transactions['product'].str.lower().replace('ё', 'е', regex=True)
````

Проверим данные на наличие пропусков.

````python
transactions.isna().sum()
````

**Вывод:** пропусков в данных нет.

Проверим наличие явных дубликатов. Проверять будем без учета даты.

````python
(
    transactions[['customer_id', 'order_id', 'product', 'price']]
    .loc[transactions[['customer_id', 'order_id', 'product', 'price']].duplicated()]
)
````

**Вывод:** таких строк оказалось 1884. Уберем дубликаты.

````python
transactions = (
    transactions
    .drop_duplicates(subset=['customer_id', 'order_id', 'product', 'price'])
    .reset_index(drop=True)
)
````

Проверим данные на аномалии. Будем исходить из того, что у одного *order_id* может быть только один *customer_id*.

````python
# посчитаем количество уникальных покупателей в каждом заказе
abnormal_orders = (
    transactions
    .groupby('order_id', as_index=False)['customer_id']
    .nunique()
    .rename(columns={'customer_id':'nunique'})
    .sort_values(by='nunique', ascending=False)
)
abnormal_orders = abnormal_orders.query('nunique > 1')
print('Абсолютное количество аномальных заказов:', abnormal_orders['order_id'].count())
print('Отностильное количество аномальных заказов: {:.2%}'.format(abnormal_orders['order_id'].count()/transactions['order_id'].count()))
````

**Вывод:** таких заказов получилось всего 29, или же 0.52% от всех данных, поэтому можно их убрать и это не сильно повлияет на анализ. Такие заказы могли получится вследствие технических ошибок на стороне сервера.

````python
# уберем строки с аномальными заказами из данных
abnormal_orders_id = abnormal_orders['order_id']
transactions = transactions.query('order_id not in @abnormal_orders_id')
````

````python
# функция ищет вхождение ключевого слова в строке
# и выдает соответствующую категорию

def get_category(df, categories):
    for key in categories.keys():
        for value in categories[key]:
            df.loc[df['product'].str.contains(value), 'category'] = key
````

Проанализируем в какие часы, дни недели и месяца происходит больше всего заказов. Сделаем общий вывод.

## Анализ основного и дополнительного ассортимента

Для начала поделим весь ассортимент на основной и дополнительный. Для этого будем проводить параллельный ABC анализ на двух критериях: выручка и количество продаж.

Предварительно подготовим функцию, которая будет возвращать к строке с товаром категорию следующим образом: 
- категория А — приоритетные товары, до 80% от общего оборота / количества товаров;
- категория В — обычные товары, от 80% до 95% общего оборота / количества товаров;
- категория С — товары-аутсайдеры, от 95% до 100% общего оборота / количества товаров (все, что осталось).

````python
def get_assortment(df):
    df.loc[(df['cum_ratio,%'] < 80), 'category'] = 'A'
    df.loc[(df['cum_ratio,%'] >= 80) & (df['cum_ratio,%'] < 95), 'category'] = 'B' 
    df.loc[(df['cum_ratio,%'] >= 95), 'category'] = 'C' 
````

Поделим товар на категории по выручке.

````python
revenue_by_product = (
    transactions
    .groupby('product', as_index=False)
    .agg({'total_price':'sum'})
    .rename(columns={'total_price':'sum'})
    .sort_values(by='sum', ascending=False)
)
revenue_by_product['ratio,%'] = revenue_by_product['sum'] / revenue_by_product['sum'].sum() * 100
revenue_by_product['cum_ratio,%'] = revenue_by_product['ratio,%'].cumsum()
get_assortment(revenue_by_product)
revenue_by_product.head()
````

Поделим товар на категории по количеству проданных товаров.

````python
sales_by_product = (
    transactions
    .groupby('product', as_index=False)
    .agg({'quantity':'sum'})
    .rename(columns={'quantity':'sales'})
    .sort_values(by='sales', ascending=False)
)
sales_by_product['ratio,%'] = sales_by_product['sales'] / sales_by_product['sales'].sum() * 100
sales_by_product['cum_ratio,%'] = sales_by_product['ratio,%'].cumsum()
get_assortment(sales_by_product)
sales_by_product.head()
````

Теперь объединим эти категории и добавим их к основному датафрейму, назовем столбец `assortment`. Добавим также столбец `assort_split`, в котором определим основной (main) это товар или дополнительный (additional).

````python
# объединяем категории
sales_by_product = \
pd.merge(sales_by_product, revenue_by_product[['product','category']], on='product')
sales_by_product['assortment'] = sales_by_product['category_x'] + sales_by_product['category_y']

# добавляем получившуюся категорию к основному датафрейму
transactions = \
pd.merge(transactions, sales_by_product[['product','assortment']], on='product')
````

````python
# определим товары по основным и дополнительным позициям
main = ["AA", "AB", "BA", "AC", "CA"]
additional = ["BB", "BC", "CB", "CC"]

transactions.loc[transactions['assortment'].isin(main), 'assort_split'] = 'main'
transactions.loc[transactions['assortment'].isin(additional), 'assort_split'] = 'additional'
transactions.head()
````

**Вывод по выручке в группах:**

- меньше всего выручки приносят группы BC 2.52%, CC 1.46% и AC 1.04%.

````python
sold_by_assort = (
    transactions
    .groupby('assortment', as_index=False)
    .agg({'quantity':'sum'})
    .rename(columns={'quantity':'sold'})
    .sort_values(by='sold', ascending=False)
)

# строим диаграмму с сегментами
fig = go.Figure(data = go.Pie(labels=sold_by_assort['assortment'],
                             values=sold_by_assort['sold'])) 
fig.update_layout(title='Число продаж товаров по группам',
                  annotations=[dict(x=1.12,
                                    y=1.05,
                                    text='Количество проданных товаров',
                                    showarrow=False)])
fig.show()
````

## Проверка гипотез

В ходе исследования мы поделили весь ассортимент магазина на основной и дополнительный. На основе этого сформируем гипотезы для проверки.

Для начала тестирования сформируем статистические гипотезы - нулевую и альтернативную:
```
H_0: Средний чек основного ассортимента = средний чек дополнительного ассортимента 
H_a: Средний чек основного ассортимента > средний чек дополнительного ассортимента 
alpha = 0.05
```

Чтобы определиться с критерием для проверки для начала выявим количество уникальных клиентов и уникальных заказов в каждой из групп, а также проверим распределение на нормальность критерием Шапиро-Уилка.

Отдельно выделим функцию с проверкой распределения на нормальность, а также односторонний t-тест.

````python
# тест Шапиро-Уилка

def shapiro_test(arr, alpha):
    result = st.shapiro(arr)

    if result.pvalue > alpha:
        print("Выборка не подчиняется закону нормального распределения")
    else:
        print("Выборка подчиняется закону нормального распределения")
````

````python
def get_ttest_result(arr1, arr2, alpha):
    results = st.ttest_ind(arr1, arr2, equal_var=False)

    print('p-значение:', results.pvalue / 2)

    if (results.pvalue / 2 < alpha) and (arr1.mean() > arr2.mean()):
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу') 
````

````python
# задаем выборки и уровень статистической значимости
alpha = 0.05

sample_main = transactions.query('assort_split =="main"') 
sample_add = transactions.query('assort_split =="additional"') 

print('''Количество уникальных клиентов основного ассортимента: {0}
Количество уникальных заказов основного ассортимента: {1}\n
Количество уникальных клиентов дополнительного ассортимента: {2}
Количество уникальных заказов дополнительного ассортимента: {3}
'''.format(sample_main['customer_id'].nunique(), sample_main['order_id'].nunique(),
     sample_add['customer_id'].nunique(), sample_add['order_id'].nunique()))
````

````python
shapiro_test(sample_main['total_price'], alpha)
````

````python
shapiro_test(sample_add['total_price'], alpha)
````

Поскольку выборки подчиняются закону нормального распределения и они достаточно большие, то можем использовать односторонний t-тест для независимых выборок (в коде возьмем двусторонний, при этом p-value поделим на 2 и добавим условие сравнение средних двух выборок).

````python
get_ttest_result(sample_main['total_price'], sample_add['total_price'], alpha)
````

**Вывод:** pvalue оказалось значительно меньше 0.05, в следствие чего мы отвергаем нулевую гипотезу, а значит у нас есть основания считать, что средний чек основного ассортимента значительно больше среднего чека дополнительного ассортимента. Из этого вытекает необходимость увеличения ассортимента товаров основного ассортимента и активного их продвижения на сайте.

Проверим также продажи основного и дополнительного ассортимента по будням и отдельно по выходным. Сделаем общий вывод.

Для начала тестирования сформируем статистические гипотезы - нулевую и альтернативную:

```
H_0: Среднее количество продаж основного ассортимента в будни = среднее количество продаж дополнительного ассортимента в будни 
H_a: Среднее количество продаж основного ассортимента в будни > среднее количество продаж дополнительного ассортимента в будни
alpha = 0.05
```

````python
# определяем выборки
workdays = [0, 1, 2, 3, 4]
sample_main_workday = sample_main.query('weekday in @workdays')['quantity']
sample_add_workday = sample_add.query('weekday in @workdays')['quantity']
````

Проверим распределение на нормальность.

````python
shapiro_test(sample_main_workday, alpha)
````

````python
shapiro_test(sample_add_workday, alpha)
````

Таким же образом проверим продажи в выходные дни.

Для начала тестирования сформируем статистические гипотезы - нулевую и альтернативную:

```
H_0: Среднее количество продаж основного ассортимента в выходные дни = среднее количество продаж дополнительного ассортимента в выходные дни

```
