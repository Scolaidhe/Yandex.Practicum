````python
data['RICE'] = (
    data['reach'] *\
    data['impact'] *\
    data['confidence'] / data['efforts']    
)
data[['hypothesis', 'RICE']].sort_values('RICE', ascending=False)
````

**Общий вывод:** при применении фреймворка `RICE` на первое место вырвалась гипотеза с номером 7, а также вместо 8-й гипотезы появилась 2-я. ## Куммулятивная выручка по группам

Построим график кумулятивной выручки по группам. Сделаем выводы и предположения.

````python
# создаем датафрейм с уникальными значениями date и group
datesGroups = orders[['date', 'group']].drop_duplicates()
# подготавливаем датафреймы для объединения
# считаем суммарную выручку заказов и количество посетителей по группам
ordersAggregated = \
datesGroups.apply(
    lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg({
        'date' : 'max',
        'group' : 'max',
        'transactionId' : pd.Series.nunique,
        'visitorId' : pd.Series.nunique,
        'revenue' : 'sum'}), axis=1).sort_values(by=['date', 'group'])

visitorsAggregated = \
datesGroups.apply(
    lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg({
        'date' : 'max',
        'group' : 'max',
        'visitors' : 'sum'
    }), axis = 1).sort_values(by=['date', 'group'])
````

````python
# объединяем обработанные датафреймы
# по колонкам date и group
cumulativeData = ordersAggregated.merge(
    visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group']
)
# переименуем колонки датафрейма
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
print(cumulativeData.head(5))
````

````python
# разделяем данные о суммарной выручке по группам
cumulativeRevenueA=cumulativeData.query('group == "A"')[['date', 'revenue', 'orders']]
cumulativeRevenueB=cumulativeData.query('group == "B"')[['date', 'revenue', 'orders']]
````

````python
# выводим график кумулятивной выручки по дням 
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')

plt.ylabel('выручка, млн')
plt.xlabel('дата')
plt.xticks(rotation=45)
plt.grid(True)
plt.title('График куммулятивной выручки по группам')
plt.legend()
plt.show()
````

**Выводы:**
- примерно с 2019-08-18 группа B сильно вырвалась вперед по сравнению с группой A по куммулятивной выручке, возможно на это повлияли аномально большие заказы;
- до 18 числа куммулятивная выручка у групп развивалась примерно схожим образом.

## Относительное изменение кумулятивного среднего чека группы B к группе A

Построим график относительного изменения кумулятивного среднего чека группы B к группе A. Сделаем выводы и предположения.

````python
# объединим по дате данные по группам в один датафрейм 
mergedCumulativeRevenue = \
cumulativeRevenueA.merge(
    cumulativeRevenueB, 
    left_on='date',
    right_on='date',
    how='left',
    suffixes=['A', 'B']
)
````

````python
plt.plot(
    mergedCumulativeRevenue['date'], 
    (mergedCumulativeRevenue['revenueB']\
    /mergedCumulativeRevenue['ordersB'])\
    /(mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA'])-1
)

plt.axhline(y=0, color='black', linestyle='--')

plt.ylabel('отношение\nкуммулятивной\nвыручки', rotation=360, labelpad=40)
plt.xlabel('дата')
plt.xticks(rotation=45)
plt.grid(True)
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A')
plt.show()
````

## Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным

Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным. Сделаем выводы и предположения.

````python
# посчитаем куммулятивное число пользователей по датам
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']
````

````python
# посчитаем куммулятивное число заказов
# и куммулятивную стоимость заказов по датам
ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]
````

````python
# объединяем получившиеся выше датафреймы в один большой по дате
data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)

#print(data.head(5))
````

````python
# отдельно выделим заказы по группам
ordersByUsersA = (
    orders[orders['group']=='A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns = ['visitorId', 'orders']

ordersByUsersB = (
    orders[orders['group']=='B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['visitorId', 'orders']
````

````python
# посчитаем отдельно по группам переменные, в которых
# пользователям с заказами будет соответствовать число заказов пользователя
# а пользователям без заказов — нули
sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
         pd.Series(
            0, index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders']
                )),
         name='orders'
     )],
    axis=0
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
         pd.Series(
            0, index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders']
                )),
         name='orders'
     )],
    axis=0
)
````

Для начала A/B тестирования сформируем гипотезы - нулевую и альтернативную:
```
H_0: Среднее количество заказов на одного посетителя группы A = среднее количество заказов на одного посетителя группы B
H_a: Среднее количество заказов на одного посетителя группы A ≠ среднее количество заказов на одного посетителя группы B
alpha = 0.05
```

За статистический критерий будем принимать непараметрический тест Манна-Уитни. Это обусловлено тем, что данный метод работает с рангами (номерами значений в упорядоченном ряду), который не учитывает сами значения, что позволит избегнуть проблему с выбросами, сильно сдвигающих параметрические результаты.

````python
# применяем тест Манна-Уитни для подсчета pvalue
print('p-value равно: {0:.5f}'.format(
    stats.mannwhitneyu(sampleA, sampleB)[1]))


print(
    'Относительный прирост среднего количества заказов на посетителя группы B к группе A: {0:.3f}'.format(
        (data['ordersPerDateB'].sum()/data['visitorsPerDateB'].sum())\
    /(data['ordersPerDateA'].sum()/data['visitorsPerDateA'].sum())-1)
     )
````

**Выводы:**
- pvalue оказалось меньше 0.05, в следствие чего мы отвергаем нулевую гипотезу;
- в результате анализа "сырых" данных можно сказать, что в среднем количестве заказов на посетителя между группами есть статистически значимые различия, при этом относительный прирост среднего количества заказов на посетителя группы B к группе A равен 13.8%

## Cтатистическая значимость различий в среднем чеке заказа между группами по «сырым» данным

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным. Сделаем выводы и предположения.

Для начала A/B тестирования сформируем гипотезы - нулевую и альтернативную:
```
H_0: Средний чек заказа группы A = средний чек заказа группы B
H_a: Средний чек заказа группы A ≠ средний чек заказа группы B
alpha = 0.05
```

````python
# применяем тест Манна-Уитни для подсчета pvalue
print('p-value равно: {0:.3f}'
      .format(
          stats.mannwhitneyu(
              orders[orders['group']=='A']['revenue'],
              orders[orders['group']=='B']['revenue']
          )[1]
        )
     )

print('Относительный прирост среднего чека группы B к группе A: {0:.3f}'
      .format(
          orders[orders['group']=='B']['revenue'].mean()\
          /orders[orders['group']=='A']['revenue'].mean()-1
      )
     ) 
````

**Выводы:**

- относительное различие средних чеков между сегментами составляет 25.9%, что могло возникнуть из-за аномалий в группе B. 

## Cтатистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным

Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным. Найдем абсолютное и относительное число аномальных пользователей. Сделаем выводы и предположения.

Для начала A/B тестирования сформируем гипотезы - нулевую и альтернативную:
```
H_0: Среднее количество заказов на одного посетителя группы A = среднее количество заказов на одного посетителя группы B
H_a: Среднее количество заказов на одного посетителя группы A ≠ среднее количество заказов на одного посетителя группы B
alpha = 0.05
```

````python
# отдельно выделим пользователей с аномальным количеством заказов
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > percentile_95_orders]['visitorId'],
        ordersByUsersB[ordersByUsersB['orders'] > percentile_95_orders]['visitorId'],
    ],
    axis=0,
)
````

````python
# также выделим пользователей с аномальной стоимостью чека
usersWithExpensiveOrders = orders[orders['revenue'] > percentile_95_revenue]['visitorId']

# объединим данные по аномальным пользователям
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
# отдельно посчитаем абсолютное и относительное число аномальных пользователей
print('Число аномальных пользователей:' ,abnormalUsers.count())
print(
    'Отношение аномальных пользователей к числу всех уникальных пользователей: {0:.3f}%'.format(
      abnormalUsers.count()/orders['visitorId'].nunique()*100)
     )
````

Абсолютное число аномальных пользователей составило 74, при этом оно составляет 7.18% среди всех уникальных пользователей.

````python
# посчитаем отдельно по группам переменные, в которых
# пользователям с заказами будет соответствовать число заказов пользователя
# а пользователям без заказов — нули
# при этом исключим аномальных пользователей

sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['visitorId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['visitorId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
````

````python
# применяем тест Манна-Уитни для подсчета p-value
print('p-value равно: {0:.5f}'
      .format(
    stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))


print(
    'Относительный прирост среднего количества заказов на посетителя группы B к группе A: {0:.3f}'
    .format(
        sampleBFiltered.mean()/sampleAFiltered.mean()-1)
     )
````

**Выводы:** 
- pvalue оказалось меньше 0.05, в следствие чего мы отвергаем нулевую гипотезу;
- в результате анализа "очищенных" данных можно сказать, что в среднем количестве заказов на посетителя между группами есть статистически значимые различия, при этом относительный прирост среднего количества заказов на посетителя группы B к группе A равен 14.8%;
- вывод оказался аналогичен результатам анализа "сырых" данных, увеличился только относительный прирост среднего количества заказов на посетителя группы B к группе A на 1%.

## Cтатистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным. Сделаем выводы и предположения.

Для начала A/B тестирования сформируем гипотезы - нулевую и альтернативную:
```
H_0: Средний чек заказа группы A = средний чек заказа группы B
H_a: Средний чек заказа группы A ≠ средний чек заказа группы B
alpha = 0.05
```

````python
print(
    'p-value равно: {0:.3f}'.format(
        stats.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )[1]
    )
)

print(
    "Относительный прирост среднего чека очищенной группы B: {0:.3f}".format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        - 1
    )
)
````

**Выводы:**
- p-value значительно увеличилось и близко к 1, значит нулевую гипотезу отвергнуть не получится;
- при этом разница между сегментами сократилась с 25.9% до 2%, что показывает как сильно аномалии влияют на тест

## Решение по результатам теста

**Общий вывод по проведенному тесту:** 
- Есть статистически значимое различие по среднему количеству заказов на посетителя между группами как по сырым данным, так и после фильтрации аномалий, в лучшую сторону здесь выделяется сегмент B с относительным приростом примерно в 14.8%.
- По сырым данным есть статистически значимое различие по среднему чеку между группами. Однако после удаления аномалий статистически значимое различие не наблюдается, что указывает на то как повлияли аномалии.
- График различия среднего количестве заказов на посетителя сообщает, что с начала теста сегмент B лидирует по метрике и фиксируется на среднем значении около 12% прироста относительно сегмента A.
- График различия среднего чека колеблется, из чего можно было сделать вывод об аномалиях в группах.

Исходя из вышесказанного, нужно **остановить тест и зафиксировать победу группы B над группой A**, по скольку в лучшую сторону здесь выделяется группа B, и даже с учетом того что нет статистически значимого различия по среднему чеку между группами, у неё лучше показатели по конверсии.
